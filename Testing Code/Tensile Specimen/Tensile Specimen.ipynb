```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def load_csv(filename):
    data = pd.read_csv(filename)
    return data
```

```python
L = 50  # Length
b = 19.9500  # Width
d = 13.1700  # Thickness

# Load CSV files
ft_data = load_csv('K-P-K 1 FT.csv')
st_data = load_csv('K-P-K 1 ST.csv')

# Calculate stress and strain
ft_data['Stress'] = ft_data['0'] / (b * d)
st_data['Strain'] = st_data['0'] / L

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(st_data['Strain'], ft_data['Stress'], linestyle='-')

plt.title('(K-P-K) Tensile Stress vs. Strain')
plt.xlabel('Strain')
plt.ylabel('Tensile Stress (σ)')
plt.grid(True)
plt.show()
```

```python
# Constants
L = 50  # Length

# Width and thickness for each specimen
specimens = [
    {'b': 19.9500, 'd': 13.1700, 'label': 'KPKFC', 'ft_file': 'K-P-K 1 FT.csv', 'st_file': 'K-P-K 1 ST.csv'},
    {'b': 22.8000, 'd': 13.5800, 'label': 'PKPFC', 'ft_file': 'P-K-P 2 FT.csv', 'st_file': 'P-K-P 2 ST.csv'},
    {'b': 22.8000, 'd': 13.5800, 'label': 'PKFC', 'ft_file': 'P-K 3 FT.csv', 'st_file': 'P-K 3 ST.csv'}
]

# Create a single plot
plt.figure(figsize=(8, 6))

for spec in specimens:
    # Load CSV files
    ft_data = load_csv(spec['ft_file'])
    st_data = load_csv(spec['st_file'])

    # Calculate stress and strain
    ft_data['Stress'] = ft_data.iloc[:, 0] / (spec['b'] * spec['d'])
    st_data['Strain'] = st_data.iloc[:, 0] / L

    # Plotting
    plt.plot(st_data['Strain'], ft_data['Stress'], linestyle='-', label=spec['label'])

plt.title('Tensile Stress vs. Strain')
plt.xlabel('Strain')
plt.ylabel('Tensile Stress (σ)')
plt.grid(True)
plt.legend()
plt.show()
```
